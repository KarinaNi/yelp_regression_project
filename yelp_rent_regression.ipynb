### Creating a common function 

I create a common function that models these features to avoid redundancy of code and I can test different features and training/testing them


````python
import numpy as np
def model_features(feature_list):

    # access features and ratings as X and y 
    features = df.loc[:, feature_list]
    ratings = df.loc[:, 'stars']

    X_train, X_test, y_train, y_test = train_test_split(features, ratings, test_size = 0.2, random_state = 1)

    if len(X_train.shape) < 2:
        X_train = np.array(X_train).reshape(-1, 1)
        X_test = np.array(X_test).reshape(-1, 1)
        
    model = LinearRegression()
    model.fit(X_train, y_train)

    
    print(sorted(list(zip(feature_list, model.coef_)), key = lambda x: abs(x[1]), reverse = True))

    y_pred = model.predict(X_test)

    plt.scatter(y_test, y_pred, alpha = 0.1)
    plt.xlabel('actual ratings')
    plt.ylabel('predicted ratings')
    plt.ylim(1,5)
    plt.show()
    
````
